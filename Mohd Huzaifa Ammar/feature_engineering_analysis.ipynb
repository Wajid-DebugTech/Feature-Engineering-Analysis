
# Feature Engineering & Impact Analysis

## Objective
Perform feature engineering on a sample dataset and evaluate its impact on model performance.

Dataset Used: Synthetic E-commerce Sales Dataset


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```

np.random.seed(42)

n = 500
data = pd.DataFrame({
    'price': np.random.randint(100, 1000, n),
    'quantity': np.random.randint(1, 10, n),
    'discount': np.random.uniform(0, 0.3, n),
    'marketing_spend': np.random.randint(1000, 5000, n)
})

data['revenue'] = (data['price'] * data['quantity']) * (1 - data['discount']) + 0.05 * data['marketing_spend']

data.head()
```

```

X = data[['price', 'quantity', 'discount', 'marketing_spend']]
y = data['revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

baseline_r2 = r2_score(y_test, y_pred)
print("Baseline R2 Score:", baseline_r2)
```

```

# New Features
data['price_after_discount'] = data['price'] * (1 - data['discount'])
data['revenue_per_unit'] = data['revenue'] / data['quantity']
data['marketing_efficiency'] = data['revenue'] / data['marketing_spend']

data.head()
```

```

X_new = data[['price', 'quantity', 'discount', 'marketing_spend',
              'price_after_discount', 'revenue_per_unit', 'marketing_efficiency']]

X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

new_r2 = r2_score(y_test, y_pred)
print("New R2 Score:", new_r2)
```

```

scores = [baseline_r2, new_r2]

plt.bar(['Before FE', 'After FE'], scores)
plt.title("Model Performance Comparison")
plt.ylabel("R2 Score")
plt.show()
```
